```python
# Question 1 - Output the percentage of people visiting the page via a mobile phone or a tablet device
# To answer this question, we have to figure out the total amount of visitors 
# and cross this with visitors via mobile and tablet
```

```python
# Let's eliminate the Date column so we can cleanly sum up values

df2 = df.drop('Date',axis=1)

df2
```

```python
# Summing values

col_list = list(df2)

df2.loc['Sum'] = df2[col_list].astype(int).sum(axis=0)

df2
```

```python
# Focus on the sum

df2 = df2.tail(1)

df2
```

```python
# Eliminate 'Total' to work with our sums only (we avoid adding sums twice)

df2 = df2.drop(['Total'], axis=1)

df2
```

```python
# Transposing

df2_tr = df2.transpose()
df2_tr
```

```python
# Creating percentages

df2_tr['Perc'] = df2_tr['Sum'].div(df2_tr['Sum'].sum())*100
df2_tr
```

```python
#Adding Mobile and Tablet percentages

(df2_tr.iloc[0]['Perc']) + (df2_tr.iloc[2]['Perc'])
```

```python
# Q1 answer: 81.3% of people visited the page via mobile phone or tablet device
```

```python
# Question 2 - Generate a text file containing a table listing the mean number 
# of Total Page Views during each hourly time period
```

```python
# First, let's organize the date

from datetime import datetime
```

```python
df['Time'] = pd.to_datetime(df['Date']).dt.time
df['Hour'] = pd.to_datetime(df['Date']).dt.hour
df['Date'] = pd.to_datetime(df['Date']).dt.date

df
```

```python
#Separating only what interests us: hours and total number of views (my first idea was grouping views by hour)

df_subset = df[['Hour', 'Total']]
df_subset
```

```python
#Finding the mean for every hour

df_subset = df_subset.rolling(4).mean() 

df_subset
```

```python
# Now, we locate the elements for every hour. Had a little trouble here. At first located every third element,
# then realized I had to start from 3rd and locate every 4th element

q2 = df_subset.iloc[3::4, :]

q2
```

```python
# Voil! Q2 answer: create text file with total page views for every hour

a_dataframe = pd.DataFrame(q2)

numpy_array = a_dataframe.to_numpy()
np.savetxt("Total_Page_Views.txt", numpy_array, fmt = "%d")
```
